# Spam Classifier 
1) load libs, and data
2) create a empty corpus
3) loop through messages, regex to remove unwanted symbils, lowercase them, split to remove spaces and form a list
4) stem them and store in the list, finally join to form a trimmed sentence
5) append the sentence to corpus
6) train test split
7) import tfidf
8) fit transform on the test with max features mentioned, and n-grams
9) transform on x_test, this way voacb in test is not leaked in tfidf or cv vectorizer
10) array is ready with bag odf words 
11) Multinomial NB - naive algo ideal for text realted task
12) predict and calculate accuracy, f1, precision etc..

```python
import numpy as np
import pandas as pd
messages = pd.read_csv('SpamClassifier-master/smsspamcollection/SMSSpamCollection',sep='\t',names=["label","message"])
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z0-9]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(x) for x in review if not x in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
ti = TfidfVectorizer(max_features=2000,ngram_range=(2,3))
```

```python
from sklearn.model_selection import train_test_split
```

```python
corpus
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,0].values
y
```

```python
X_tr,X_test,y_tr,y_test = train_test_split(corpus,y,test_size = 0.25)
```

```python
X_tr = ti.fit_transform(X_tr).toarray()
X_test = ti.transform(X_test).toarray()
```

```python
X_tr
```

```python
X_test
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
ti.vocabulary_
```

```python
tfidf_pred_model = MultinomialNB().fit(X_tr,y_tr)
```

```python
y_pred = tfidf_pred_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_pred=y_pred,y_true=y_test)
```

```python
print(classification_report(y_pred=y_pred,y_true=y_test))
```
